### Contour clipping

```python
import os
import numpy as np
import rasterio
from rasterio.mask import mask

# Output folder base
output_base = "/content/drive/MyDrive/data_set/climate/algeria_tunisia_climate/"
os.makedirs(output_base, exist_ok=True)

# Create subfolders for each variable
folders = {var: os.path.join(output_base, var.lower()) for var in climate_data.keys()}
for folder in folders.values():
    os.makedirs(folder, exist_ok=True)

# Dictionary to store clipped rasters in memory
clipped_rasters = {var: [] for var in climate_data.keys()}

# Loop over each variable
for var_name, file_list in climate_data.items():
    folder = folders[var_name]

    for i, tif in enumerate(file_list, 1):
        with rasterio.open(tif) as src:
            # Clip raster to Algeria + Tunisia
            out_image, out_transform = mask(src, gdf.geometry, crop=True)

            # Mask nodata
            data = np.ma.masked_equal(out_image[0], src.nodata)

            # Save to Drive
            out_meta = src.meta.copy()
            out_meta.update({
                "driver": "GTiff",
                "height": out_image.shape[1],
                "width": out_image.shape[2],
                "transform": out_transform,
                "nodata": src.nodata
            })

            out_path = os.path.join(folder, f"{var_name}_{i:02d}_clipped.tif")
            with rasterio.open(out_path, "w", **out_meta) as dest:
                dest.write(out_image)

            # Store in memory
            clipped_rasters[var_name].append({
                "data": data,
                "transform": out_transform,
                "meta": out_meta,
                "filename": out_path
            })

            print(f"Saved {out_path} and stored in memory.")
```

```python
import rasterio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# === Paths ===
raster_path = "/content/drive/MyDrive/data_set/soil_type/HWSD_Algeria_Tunisia.tif"
csv_path = "/content/drive/MyDrive/data_set/soil_type/soil_dataset.csv"

# === Load raster ===
raster = rasterio.open(raster_path)
raster_data = raster.read(1)
raster_nodata = raster.nodata
raster_data = np.where(raster_data == raster_nodata, np.nan, raster_data)


# === Load CSV ===
df = pd.read_csv(csv_path)

# Ensure ID column is numeric and consistent
df['ID'] = df['ID'].astype(int)

# Exclude non-attribute columns
excluded = ['ID', 'geometry', 'joinMitray']  # add any others you want to skip
attributes = [c for c in df.columns if c not in excluded]


# Build lookup from raster ID -> dataframe index (for speed)
id_to_idx = pd.Series(df.index.values, index=df['ID']).to_dict()

# quick raster ID sample (for debugging)
raster_ids_sample = np.unique(raster_data[~np.isnan(raster_data)])[:20]


# plotting grid
n = len(attributes)
cols = 2
rows = (n + cols - 1) // cols
plt.figure(figsize=(16, 5 * rows))

# For diagnostics collect problems
diagnostics = {}

for i, attr in enumerate(attributes, start=1):
    print(f"\n--- plotting attribute: {attr} ---")

    # Skip columns that are all nulls
    if df[attr].isna().all():
        print(f"Skipping {attr}: all NaN")
        continue

    # Normalize categorical values in dataframe copy (but keep original)
    is_numeric_col = np.issubdtype(df[attr].dtype, np.number)

    # For categorical columns, create normalized strings
    if not is_numeric_col:
        df_attr_norm = df[attr].astype(str).where(~df[attr].isna(), None)
        df_attr_norm = df_attr_norm.apply(lambda x: x.strip() if isinstance(x, str) else x)
        # optional: unify case (uncomment if helpful)
        # df_attr_norm = df_attr_norm.apply(lambda x: x.lower() if isinstance(x, str) else x)
        unique_vals = pd.Series(df_attr_norm.dropna().unique())
        unique_vals = unique_vals.tolist()
        print(f"Categorical unique count (non-null): {len(unique_vals)}")
    else:
        # numeric column: use original
        df_attr_norm = df[attr]
        unique_vals = None

    # prepare result arrays
    result_num = np.full_like(raster_data, np.nan, dtype='float32')
    result_cat = None

    # fill numeric or prepare categorical integer-coded array
    # iterate only over unique raster ids present
    mask = ~np.isnan(raster_data)
    raster_unique_ids = np.unique(raster_data[mask])
    raster_unique_ids = raster_unique_ids[~np.isnan(raster_unique_ids)]

    unmatched_ids = []
    mapped_count = 0

    if is_numeric_col:
        # numeric: assign float values
        for rid in raster_unique_ids:
            rid_int = int(rid)
            if rid_int in id_to_idx:
                val = df.at[id_to_idx[rid_int], attr]
                if pd.isna(val):
                    continue
                try:
                    result_num[raster_data == rid] = float(val)
                    mapped_count += 1
                except Exception as e:
                    # cannot convert -> skip
                    unmatched_ids.append((rid_int, val, str(e)))
            else:
                unmatched_ids.append((rid_int, None, "ID not in df"))
        has_data = np.isfinite(result_num).any()
    else:
        # categorical: build mapping from normalized value -> integer
        val_list = [str(v).strip() for v in unique_vals if pd.notna(v)]
        val_to_int = {v: idx for idx, v in enumerate(val_list)}
        result_cat = np.full_like(raster_data, np.nan, dtype='float32')

        for rid in raster_unique_ids:
            rid_int = int(rid)
            if rid_int in id_to_idx:
                raw_val = df.at[id_to_idx[rid_int], attr]
                if pd.isna(raw_val):
                    unmatched_ids.append((rid_int, None, "value is NaN"))
                    continue
                norm_val = str(raw_val).strip()
                # if using lowercase normalization uncomment both lines:
                # norm_val = norm_val.lower()
                if norm_val in val_to_int:
                    result_cat[raster_data == rid] = val_to_int[norm_val]
                    mapped_count += 1
                else:
                    unmatched_ids.append((rid_int, raw_val, "no matching normalized category"))
            else:
                unmatched_ids.append((rid_int, None, "ID not in df"))
        has_data = np.isfinite(result_cat).any()

    print(f"Mapped unique raster IDs -> attribute values: {mapped_count}")
    if unmatched_ids:
        print(f"Unmatched examples (up to 10): {unmatched_ids[:10]}")

    # Save diagnostics
    diagnostics[attr] = {
        "mapped_count": mapped_count,
        "unmatched_examples": unmatched_ids[:10],
        "has_data": bool(has_data)
    }

    # If nothing mapped, print helpful debug and continue
    if not has_data:
        print(f" No mapped cells for attribute {attr}. Skipping plot.")
        continue

    # Plotting
    ax = plt.subplot(rows, cols, i)
    if is_numeric_col:
        img = ax.imshow(result_num, cmap='viridis')
        cb = plt.colorbar(img, ax=ax, fraction=0.046, pad=0.04)
        cb.set_label(attr)
    else:
        # categorical plotting with colorbar labels
        cmap = ListedColormap(plt.cm.tab20.colors[:len(val_list)])
        img = ax.imshow(result_cat, cmap=cmap)
        cb = plt.colorbar(img, ax=ax, fraction=0.046, pad=0.04)
        # set ticks and labels
        ticks = np.arange(len(val_list)) + 0.5  # center ticks
        cb.set_ticks(np.arange(len(val_list)))
        cb.set_ticklabels(val_list)
        cb.ax.tick_params(labelsize=8)
        cb.set_label(attr)

    ax.set_title(attr)
    ax.axis('off')

plt.tight_layout()
plt.show()


```

```python
import geopandas as gpd
import pandas as pd
import rasterio
from rasterio.mask import mask
from rasterio.plot import show
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource

# ---------------------------
# Step 1: Load country boundaries
# ---------------------------
gdf = gpd.read_file("/content/drive/MyDrive/data_set_old/kontur/contour.geojson")

# Vrifier le systme de coordonnes
# Il faut qu'il corresponde  celui du raster
print("CRS du contour :", gdf.crs)

# ---------------------------
# Step 2: Open raster and clip
# ---------------------------
raster_path = "/content/drive/MyDrive/data_set_old/Elevation/be15_grd/be15_grd/w001001.adf"

with rasterio.open(raster_path) as src:
    print("CRS du raster :", src.crs)

    # Si les CRS diffrent, reprojeter le GeoDataFrame
    if gdf.crs != src.crs:
        gdf = gdf.to_crs(src.crs)

    # Convertir en liste de gomtries (GeoJSON)
    geoms = [feature["geometry"] for feature in gdf.__geo_interface__["features"]]

    # Clip raster
    out_image, out_transform = mask(src, geoms, crop=True)
    out_meta = src.meta.copy()

# ---------------------------
# Step 3: Update metadata
# ---------------------------
out_meta.update({
    "driver": "GTiff",
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform,
    "count": 1
})

# ---------------------------
# Step 4: Save clipped raster
# ---------------------------
out_raster_path = "/content/drive/MyDrive/data_set/Elevation/clipped_algeria_tunisia.tif"
with rasterio.open(out_raster_path, "w", **out_meta) as dest:
    dest.write(out_image)

print(" Clipped raster saved to:", out_raster_path)

# ---------------------------
# Step 5: Optional visualization
# ---------------------------
with rasterio.open(out_raster_path) as clipped:
    plt.figure(figsize=(8,6))
    show(clipped, cmap='terrain')
    plt.title("Clipped Elevation - Algeria & Tunisia")
    plt.show()
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd

# === Load Algeria + Tunisia contours again for plotting ===

gdf_both = gpd.read_file("/content/drive/MyDrive/data_set/kontur/contour.geojson")

# === Load clipped raster ===
with rasterio.open("/content/drive/MyDrive/data_set/Elevation/clipped_algeria_tunisia.tif") as src:
    data = src.read(1)
    data = np.ma.masked_equal(data, src.nodata)
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

# === Plot clipped raster with borders ===
fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(data, cmap='turbo', extent=extent, origin='upper')
gdf_both.boundary.plot(ax=ax, edgecolor='black', linewidth=1)
ax.set_title("Elevation  Algeria + Tunisia", fontsize=14, fontweight='bold')
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
plt.colorbar(im, ax=ax, label="Soil Type Code")
plt.tight_layout()
plt.show()
```
